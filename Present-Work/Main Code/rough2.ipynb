```python
textures=['3_Bumps','6_Bumps','4_Ridges','6_Ridges','4_Waves','6_Waves','4_Blob','6_Blob','Plain']
iterations=[1,2]
ans=0
# for k in np.arange(1,200,0.5):
D={}
z_f=[]
class_f=[]
pass_f=[]
iter_f=[]
v_f=[]
for i in textures:
    for j in iterations:
        for p in range(0,8):
            for l in range(1,17):
                map = mapPoints(i)  
                pointmap = map.pointMap(j)
                for key in pointmap.keys():
                    for k in range(len(pointmap[key])):
                        pointmap[key][k] = np.array(pointmap[key][k])
                    pointmap[key] = np.array(pointmap[key],dtype=object)
                v = pointmap[str(l)][p][:,3]
                v=list(v)
                v_f.append(v)
                # v = np.abs(v*20 - 2.44*19 - 2.5)+1
                z = pointmap[str(l)][p][:,2]
                z=list(z)
                z_f.append(z)
                Class=[i]*len(z)
                class_f.append(Class)
                pass_no=[p]*len(z)
                pass_f.append(pass_no)
                iter_no=[j]*len(z)
                iter_f.append(iter_no)

```
